### Ejercicio 4 
Atributos Privados
**Objetivo:** Proteger datos con atributos privados
```python
 Crea una clase "CajaFuerte" con:
 - Atributo privado: __dinero (inicia en 0)
 - M√©todo depositar(cantidad): agrega dinero (solo si cantidad > 0)
 - M√©todo retirar(cantidad): retira dinero (solo si hay suficiente)
 - M√©todo ver_saldo(): retorna el saldo

````python
class CajaFuerte:
    def __init__(self):
        self.__dinero = 0

    def depositar(self, cantidad):
        if cantidad > 0:
            self.__dinero += cantidad
    
    def retirar(self, cantidad):
        if self.__dinero >= cantidad:
            self.__dinero -= cantidad
            return print(f"Has retirado {cantidad} de tu cuenta")
        
    def ver_saldo(self):
        return self.__dinero
    
caja = CajaFuerte()
caja.depositar(1000)
caja.retirar(300)
print(caja.ver_saldo())
````

### Exercise 6
M√âTODOS ESPECIALES Y DE CLASE
**Objetivo:** Personalizar la representaci√≥n del objeto
```python
 Crea una clase "Producto" con:
 - Atributos: nombre, precio, stock
 - __str__: retorna formato legible "Nombre: ‚Ç¨Precio (Stock unidades)"
 - __repr__: retorna formato t√©cnico "Producto('nombre', precio, stock)"
 - __len__: retorna el stock

 Ejemplo:
 p = Producto("Laptop", 899.99, 5)
 print(p)  # Laptop: ‚Ç¨899.99 (5 unidades)
 print(repr(p))  # Producto('Laptop', 899.99, 5)
 print(len(p))  # 5

````python
class Producto:
    def __init__(self,nombre,precio,stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

    def __str__(self):
        return f"{self.nombre}: {self.precio} con ({self.stock})"
    
    def __repr__(self):
        return f"Producto({self.nombre}, {self.precio}, {self.stock})"
    
    def __len__(self):
        return self.stock
    
p = Producto("Laptop", 899.99, 5)
print(p)  
print(repr(p))  
print(len(p)) 
````

### exercise 7

Herencia con super()
**Objetivo:** Extender comportamiento del padre
```python
 Crea:
 - Clase "Vehiculo": marca, modelo
   * M√©todo info(): retorna "Marca Modelo"
 
 - Clase "Coche" hereda de Vehiculo: puertas
   * Sobrescribe info() usando super() para a√±adir n√∫mero de puertas
 
 - Clase "Moto" hereda de Vehiculo: cilindrada
   * Sobrescribe info() usando super() para a√±adir cilindrada

````python
class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def info(self):
        return print(f"Es un {self.marca} modelo {self.modelo}")

class Coche(Vehiculo):
    def __init__(self, marca, modelo, puertas):
        self.puertas = puertas
        super().__init__(marca, modelo)

    def info(self):
        return print(f"Es un {self.marca} modelo {self.modelo} con {self.puertas} puertas")        
    
class Moto(Vehiculo):
    def __init__(self, marca, modelo, cilindrada):
        self.cilindrada = cilindrada
        super().__init__(marca, modelo)

    def info(self):
        return print(f"Con {self.cilindrada} cilindraje")
    
c = Coche("Toyota", "Corolla", 4)
print(c.info())  # Toyota Corolla - 4 puertas
````

### exercise 8
Composici√≥n
**Objetivo:** Objetos que contienen otros objetos
```python
    Crea:
    - Clase "Motor": caballos
    * M√©todo arrancar(): imprime "Motor arrancado"
 
 - Clase "Coche": marca, motor (objeto Motor)
   * Constructor crea su propio motor
   * M√©todo encender(): llama a motor.arrancar()

````python
class Motor:
    def __init__(self, caballos):
        self.caballos = caballos
    
    def arrancar(self):
        return f"motor de {self.caballos} arrancado"
    
class Coche:
    def __init__(self, marca, caballos):
        self.marca = marca
        self.motor= Motor(caballos)

    def encender(self):
        return Motor.arrancar()

coche = Coche("Ford", 150)
print(coche.encender())
````
