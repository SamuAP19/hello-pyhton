## Exercise 1: Creating Objects and Methods

Create a Python class called `Rectangle` with attributes `length` and `width`. Implement methods to calculate the area and perimeter of the rectangle. Then, create two `Rectangle` objects and demonstrate the use of these methods.

```python
class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width    
    def perimeter (self):
        return self.length * 2 + self.width * 2 
    
x = Rectangle(int(input("enter the length")), int(input("give the width")))
print(f"the area is {x.area()} ")
print(f"the perimeter is {x.perimeter()}")    

```

## Exercise 2: Access Modifiers

Define a class `Person` with private attributes `name` and `age`. Implement getter and setter methods for these attributes. Create an instance of the `Person` class and demonstrate how to set and retrieve the person’s name and age.

```python
class Person:
    def __init__(self,age, name):
        self.__age = age
        self.__name = name
    def getname(self):
        return self.__name
    def setname(self, new_name):
        self.__name = new_name
        return self.__name
    def getage(self):
        return self.__age
    def setage(self, new_age):
        self.__age = new_age
        return self.__age

p = Person( 52, "samuel")
print(p.getage())
print(p.setage(18))

    
    
```

## Exercise 3: Inheritance

Create a base class `Vehicle` with attributes `make` and `model`. Derive two classes, `Car` and `Motorcycle`, from `Vehicle`. Each derived class should have a method that displays information about the vehicle, including make and model. Create instances of both `Car` and `Motorcycle` and display their information.

```python
class Vehicle:
    def __init__(self,make,model):
         self.make = make
         self.model = model

class Car(Vehicle):
    def __init__(self, make, model, year):
        super().__init__(make, model)
        self.year = year
    def infocar(self):
        print(f"info of the car : the year is {self.year}, the model is {self.model}, the make is {self.make}")

class Moto(Vehicle):
    def __init__(self, make, model,year):
        super().__init__(make, model)
        self.year = year
    def infomoto(self):
        print(f"info of the moto : the year is {self.year}, the model is {self.model}, the make is {self.make}")

car = Car("asian","corolla"  ,2025)
moto = Moto("japan", "ninja", 2025)
car.infocar()
moto.infomoto()
    

```
