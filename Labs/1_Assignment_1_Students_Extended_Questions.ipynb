## Exercise 1

Write a Python program that prints the following pattern:

```
____A____
___B_C___
__D_E_F__
_G_H_I_J_
K_L_M_N_O
```

**Answer (write your code in the Python cell below):**

````python
rows = 5
ch = ord('A') 

for i in range(rows):       
    print("_" * (rows - i - 1), end="")     
    for j in range(i + 1):
        print(chr(ch), end="")
        ch += 1
        if j != i:  
            print("_", end="")
    print("_" * (rows - i - 1))
````

## Exercise 16

Write a program that reads a number **N** and prints a **pyramid** of `*`.

Example for N = 5:
```
    *
   ***
  *****
 *******
*********
```

````python
n = int(input("Enter the number of rows and stars"))

for x in range (1, n + 1):
    stars = 2 * x - 1
    width = 2 * n - 1
    total = stars * "*"
    print (total.center(width))
````

## Exercise 17

Write a program that reads a number **N** and prints a **diamond** of `*`.

Example for N = 3:
```
  *
 ***
*****
 ***
  *
```

````python
N = int(input("Enter a number for the diamond size: "))

width = 2 * N - 1

for i in range(1, N + 1):
  stars_count = 2 * i - 1
  stars = '*' * stars_count
  print(stars.center(width))

for i in range(N - 1, 0, -1):
  stars_count = 2 * i - 1
  stars = '*' * stars_count
  print(stars.center(width))
````
