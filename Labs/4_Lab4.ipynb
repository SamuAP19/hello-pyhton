### Exercise 6: List Flattening
Given `nested = [[1,2],[3,4],[5,[6,7]]]`, flatten the list into `[1,2,3,4,5,6,7]` recursively.

```python

nested = [[1,2],[3,4],[5,[6,7]]]
def flat(x):
    result  = []
    for item in x:
        if isinstance(item, list):
            result.extend(flat(item))
        else:
            result.append(item)
    return result

print(flat(nested))
        

        

```

```python
x = [[1,2],[3,4],[5,[6,7]]]

def fl(nested):
    result = []
    for item in nested:
        if isinstance(item, list):
            result.extend(fl(item))
        else:
            result.append(item)
    return result

print(fl(x))
```

### Exercise 10: Inverting a Dictionary
Invert the dictionary `{ 'a': 1, 'b': 2, 'c': 1 }` so values become keys and keys are stored in lists.

```python
dictionary = { 'a': 1, 'b': 2, 'c': 1 }
new_dictionary = {}
for k, v in dictionary.items():
    new_dictionary.setdefault(v,[]).append(k)

print(new_dictionary)
```

### Exercise 11: Matrix Transpose with Lists
Transpose a 3x2 matrix `[[1,2],[3,4],[5,6]]` without using NumPy.

```python
mat = [[1,2],[3,4],[5,6]]
transpose = []
new_mat = []
for y in range(2):
    for x in range(len(mat)):
        new_mat.append(mat[x][y])
    transpose.append(new_mat)

print(transpose)
```

### Exercise 12: Most Common Word
Given `text = 'data science at MIT is science and data'`, find the most frequent word.

```python
text = 'data science at MIT is science and data'
new_text = text.split()
dic = {}
for x in new_text:
    if x in dic:
        dic[x] += 1
    else:
        dic[x] = 1
org_dic = dict(sorted(dic.items(), key=lambda x:x[1], reverse=True))
print(f"the most repeated word is {next(iter(org_dic))}")


```

### Exercise 13: Anagram Groups
Group words `['listen','silent','enlist','rat','tar','art']` into anagrams using a dictionary.

```python
words = ['listen','silent','enlist','rat','tar','art']
anagrams = {}
for w in words:
    key = ''.join(sorted(w))
    anagrams.setdefault(key, []).append(w)
print(list(anagrams.values()))
```

### Exeercise 16: Patterns
Write a Python function that checks if two words given as input have the same pattern of characters or not. Two strings have the same pattern of characters if:
- They have the same length
- If two characters in the first string are equal if and only if the characters in the corresponding positions in the second string are also equal
Note: You are allowed to use strings only. No lists, no tuples, no dictionaries


```python
def same_pattern(s1,s2):
    if len(s1) != len (s2):
        return False
    for i in range(len(s1)-1):
        for j in range(i+1,len(s1)):
            if (s1[i] == s1[j]) != (s2[i] ==s2[j]):
                return False
    return True

same_pattern("abba","xyyx")

```

### Exercise 17: Removing Duplicates

Given a sorted list L, write a function that removes the duplicates from a list L.

- Use while loops only

### Exercise 18: 
From a list of integers L, find two numbers that add up to a given target S.

```python
def find_sum(l,s):
    i,j = 0, len(l)-1
    while i < j:
        if l[i] + l[j] == s:
            return l[i],l[j]
        elif l[i] + l[j]  < s:
            i +=1
        else:
             j += 1

find_sum((2,5,7,9,10,11), 16)


```
