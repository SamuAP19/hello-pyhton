### Exercise 17: Simple Calculator

Write a calculator program that asks the user for two numbers and an operator (+, -, *, /). Print the result.

````python
try:

  num1 = float(input("Introduce el primer número: "))
  operador = input("Introduce un operador (+, -, *, /): ")
  num2 = float(input("Introduce el segundo número: "))

  
  if operador == '+':
    resultado = num1 + num2
    print(f"El resultado es: {resultado}")
  elif operador == '-':
    resultado = num1 - num2
    print(f"El resultado es: {resultado}")
  elif operador == '*':
    resultado = num1 * num2
    print(f"El resultado es: {resultado}")
  elif operador == '/':
    
    if num2 == 0:
      print("Error: No se puede dividir por cero.")
    else:
      resultado = num1 / num2
      print(f"El resultado es: {resultado}")
  else:
    
    print("Operador no válido. Por favor, usa +, -, * o /.")

except ValueError:
  print("Error: Por favor, introduce números válidos.")
````

### Exercise 19: Count Vowels

Write a program that counts how many vowels (a, e, i, o, u) are in a user-entered string.

````python
user_string = input("Enter a string: ")

vowels = "aeiou"

vowel_count = 0

for char in user_string.lower():
  if char in vowels:
    vowel_count += 1

print(f"The number of vowels is: {vowel_count}")
````

### Exercise 20: Advanced Main Program

Create a program with two functions: `square(n)` and `cube(n)`. In the `main` block, ask the user for a number and print its square and cube.

````python
def square(n):
  return n ** 2

def cube(n):
  return n ** 3

if __name__ == '__main__':
  try:
    
    user = float(input("Enter a number: "))

    
    square_result = square(user)
    cube_result = cube(user)
    
    
    print(f"The square of {user} is: {square_result}")
    print(f"The cube of {user} is: {cube_result}")

  except ValueError:
    print("Invalid input. Please enter a number.")
````
