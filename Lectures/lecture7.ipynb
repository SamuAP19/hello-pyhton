```python
import csv
valid_grades = []
with open("grades.csv") as Kids:
    reader = csv.DictReader(Kids)
    for row in reader:
        try:
            Grade = int(row[' Grade'])
            valid_grades.append(Grade)
        except ValueError:
            continue

    if valid_grades:
        average = sum(valid_grades)/len(valid_grades)
        print(f"the average is {average:.2f}")
    else:
        print("No values")
        
                
        

                
```
