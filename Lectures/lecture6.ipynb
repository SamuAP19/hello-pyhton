```python
import csv
with open('./students.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        name = row['name']
        age = int(row['age'])
        gender = row['gender']
        print(f"{name} is {age} years old, is a {gender}")
```
